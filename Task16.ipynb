Q1. Load the insurance dataset using pandas and display the first 10 rows of the data.

```python
import pandas as pd
df = pd.read_csv("/insurance.csv")
print(df.head(10))
```

Q2. Use pandas functions to get basic information about the dataset: Shape of the data (shape), Column names and data types (info()), Statistical summary of numerical columns (describe())

```python
print("Shape of the data:- ")
print(df.shape)

print("\nColumn names and data types:- ")
print(df.info())

print("\nStatistical summary of numrical columns:- ")
print(df.describe())
```

Q3. Check if the dataset contains any null/missing values. Show the count of null values for each column.

```python
print("Missing values in each column:-")
print(df.isnull().sum())
```

Q4. Identify numerical and categorical columns in the dataset. Print them separately.

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

categorical_columns = df.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(numerical_columns)

print("\nCategorical Columns:")
print(categorical_columns)
```

Q5. Create distribution plots for all numerical columns (age, bmi, children, charges): Use sns.histplot() or sns.distplot() ,Create one plot per column (total 4 plots), Add meaningful titles and labels

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
sns.histplot(df["age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.subplot(2, 2, 2)
sns.histplot(df["bmi"])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")

plt.subplot(2, 2, 3)
sns.histplot(df["children"])
plt.title("Children Distribution")
plt.xlabel("Children")
plt.ylabel("Frequency")

plt.subplot(2, 2, 4)
sns.histplot(df["expenses"])
plt.title("Expenses Distribution")
plt.xlabel("Expenses")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Q6. Create count plots for all categorical columns (sex, smoker, region): Use sns.countplot() ,for each column ,Add titles and labels for clarity

```python
plt.figure(figsize=(10,8))

plt.subplot(2, 2, 1)
sns.countplot(x="sex", data=df)
plt.title("Sex Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")

plt.subplot(2, 2, 2)
sns.countplot(x="smoker", data=df)
plt.title("Smoker Distribution")
plt.xlabel("Smoker")
plt.ylabel("Count")

plt.subplot(2, 2, 3)
sns.countplot(x="region", data=df)
plt.title("Region Distribution")
plt.xlabel("Region")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

Q7. Create a heatmap to show the correlation between numerical variables. Use sns.heatmap(), Annotate the values and use a suitable color palette

```python
corr=df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm") #red = positive corr, blue = negative corr

plt.title("Correlation Heatmap")
plt.show()
```

Q8. Perform the following analysis on the charges column: Find the average insurance charges,Find the maximum and minimum charges, Group by smoker and compare the average charges for smokers vs nonsmokers

```python
print("Average Insurance Expenses:-",df["expenses"].mean())
print()

print("\nMaximum Expenses:-",df["expenses"].max())
print()

print("\nMinimum Expenses:-",df["expenses"].min())
print()

print("\nAverage Expenses by Smoker:-",)
print(df.groupby("smoker")["expenses"].mean())
```

Q9. Create a boxplot to show the distribution of charges with respect to smoker and sex. (You can create two separate boxplots or use hue parameter)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

sns.boxplot(x="sex", y="expenses", hue="smoker", data=df)

plt.title("Expenses by Sex and Smoker")
plt.xlabel("Sex")
plt.ylabel("Expenses")

plt.show()
```
