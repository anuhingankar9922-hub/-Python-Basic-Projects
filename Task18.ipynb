Q1. (Data Loading & Initial Analysis) Load the Ford Car Dataset using pandas. Display the first 10 rows and last 5 rows. Check the shape (number of rows and columns) and data types of all columns. Write observations about the dataset structure in comments.

```python
import pandas as pd
df = pd.read_csv("ford_car_dataset.csv")

print("__________::___________________:First 10 Rows:___________________::__________\n")
print(df.head(10))

print("\n_______::___________________:Last 5 Rows:___________________::_______\n")
print(df.tail())

print("\n_:Shape of Dataset:_\n")
print(df.shape)

print("\n:_____:Data Types:_____:\n")
print(df.dtypes)

print("\n:_____________:Dataset Info:____________:\n")
print(df.info())

# Observations:
# 1. The dataset contains information about Ford cars.
# 2. Each row represents one car record.
# 3. The shape shows the total number of rows and columns.
# 4. The data contains both numerical and categorical columns.
# 5. The data types help identify which columns are integers, floats, or objects (text).
# 6. The dataset appears suitable for further data cleaning and analysis.
```

Q2.(Missing & Duplicate Values) Perform a detailed check on the dataset: Find the total number of missing values in each column. Identify and remove duplicate rows if any. Write a summary of your findings and how you handled them in comments.

```python
print("Missing Values in Each Column:")
print(df.isnull().sum())

print("\nNumber of Duplicate Rows:")
print(df.duplicated().sum())

df = df.drop_duplicates()

print("\nShape After Removing Duplicates:")
print(df.shape)

# Summary:
# 1. Checked all columns for missing values using isnull().sum().
# 2. Checked duplicate rows using duplicated().sum().
# 3. Removed duplicate rows using drop_duplicates().
# 4. The cleaned dataset is now ready for further analysis.
# 5. If no missing values or duplicates are found, no changes are made to the dataset.
```

Q3. (Statistical Summary) Generate the statistical summary of all numeric columns using describe(). Identify minimum, maximum, mean, and median values for key columns like price, mileage, and year.

```python
print("____:____:____:____:____:____:Statistical Summary:____:____:____:____:____:____\n")
print(df.describe())

print("\n--:--:--:Price Statistics:--:--:--\n")
print("Minimum Price :", df["price"].min())
print("Maximum Price :", df["price"].max())
print("Mean Price    :", df["price"].mean())
print("Median Price  :", df["price"].median())

print("\n---::---Mileage Statistics---::---\n")
print("Minimum Mileage :", df["mileage"].min())
print("Maximum Mileage :", df["mileage"].max())
print("Mean Mileage    :", df["mileage"].mean())
print("Median Mileage  :", df["mileage"].median())

print("\n:-:-:-:-:-Year Statistics-:-:-:-:-:")
print("Minimum Year :", df["year"].min())
print("Maximum Year :", df["year"].max())
print("Mean Year    :", df["year"].mean())
print("Median Year  :", df["year"].median())
```

Q4.(Histogram of Numeric Features) Plot histograms for all important numeric columns (price, mileage, year, engineSize, mpg). Analyze the distribution for each feature and note your findings in comments.

```python
import matplotlib.pyplot as plt

cols = ["price", "mileage", "year", "engineSize", "mpg"]

plt.figure(figsize=(12, 8))

for i, col in enumerate(cols, 1):
    plt.subplot(3, 2, i)
    plt.hist(df[col], bins=20, edgecolor="black")
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Observations:
# 1. Price: Shows how car prices are distributed and whether most cars are low or high priced.
# 2. Mileage: Indicates the range of mileage values and whether most cars have low or high mileage.
# 3. Year: Shows the distribution of manufacturing years and the availability of newer vs older cars.
# 4. EngineSize: Most cars usually have smaller engine sizes, with fewer large-engine vehicles.
# 5. MPG: Displays the fuel efficiency distribution of the cars.
# 6. Histograms help identify the spread, skewness, and possible outliers in numerical features.
```

Q5. (Count Plots of Categorical Features) Create count plots for all categorical columns (fuelType, transmission, model, etc.) using seaborn. Analyze which categories are most common.Write insights about market trends based on these plots in comments.

```python
import seaborn as sns

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.countplot(x="fuelType", data=df)
plt.title("Fuel Type")

plt.subplot(2,2,2)
sns.countplot(x="transmission", data=df)
plt.title("Transmission")

plt.subplot(2,2,3)
sns.countplot(y="model", data=df, order=df["model"].value_counts().index)
plt.title("Car Models")

plt.tight_layout()
plt.show()

# Comments:
# 1. The count plots show the frequency of each category.
# 2. Petrol and Diesel are the most common fuel types.
# 3. Manual and Automatic are the most common transmission types.
# 4. Some Ford models appear more frequently than others, indicating higher popularity.
# 5. The plots help understand customer preferences and current market trends.
```

Q6. (Correlation Heatmap) Create a heatmap of the correlation matrix for numeric features. Use seaborn with proper annotations. Identify which features are highly correlated with the target variable (price). Write your observations in comment.

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

print("Correlation with Price:")
print(corr["price"].sort_values(ascending=False))

# Observations:
# 1. The heatmap shows the correlation between all numeric features.
# 2. Correlation values range from -1 to +1.
# 3. Values close to +1 indicate a strong positive correlation.
# 4. Values close to -1 indicate a strong negative correlation.
# 5. Features with higher positive correlation with price have a greater influence on car price.
# 6. Features with negative correlation tend to decrease as price increases.
```

Q9. (Feature Scaling) Apply Standard Scaling (using StandardScaler from sklearn) on the numeric independent features. Show the first 5 rows of scaled data.

```python
from sklearn.preprocessing import StandardScaler

df = pd.get_dummies(df, columns=["fuelType", "transmission", "model"])

X = df.drop("price", axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("______:______:_______:First 5 Rows of Scaled Data:________:______:______")
print(X_scaled.head())
```

Q10.( Mini Project - Complete Preprocessing Pipeline) Create a complete preprocessing pipeline for the Ford Car Dataset: Load and clean the data (handle missing & duplicate values). Perform EDA (histograms + count plots + heatmap). Identify input and output features. Encode categorical variables.Scale the numeric features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("ford_car_dataset.csv")

print(df.isnull().sum())

df = df.drop_duplicates()

df[["price", "mileage", "year", "engineSize", "mpg"]].hist(figsize=(10,8))
plt.tight_layout()
plt.show()

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.countplot(x="fuelType", data=df)
plt.title("Fuel Type")

plt.subplot(1,3,2)
sns.countplot(x="transmission", data=df)
plt.title("Transmission")

plt.subplot(1,3,3)
sns.countplot(y="model", data=df, order=df["model"].value_counts().index)
plt.title("Model")

plt.tight_layout()
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

X = df.drop("price", axis=1)
y = df["price"]

print("\n______:_________:_______:___Independent Features___:_______:_________:______")
print(X.columns)

print("\n-Dependent Feature-")
print(y.name)

df_encoded = pd.get_dummies(df, columns=["model", "fuelType", "transmission"])

X = df_encoded.drop("price", axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("\n--------:-----:------:First 5 Rows of Scaled Data:------:-----:--------")
print(X_scaled.head())
```
