```python
print("______________Sepal Length Column_____________\n")
print(df["sepal_length"])
```

```python
print("__________Two Columns__________\n")
print(df[["sepal_length", "petal_length"]])
```

```python
print("___________________Flowers with Sepal Length > 6_____________________\n")
print(df[df["sepal_length"] > 6])
```

```python
df["Total_Length"] = df["sepal_length"] + df["petal_length"]
```

```python
print("______________________________New Column Added_______________________________\n")
print(df.head())
```

```python
df.loc[0, "species"] = "updated_species"
```

```python
print("______________________________Updated First Row______________________________\n")
print(df.head())
```

```python
print("________________________Sorted by Sepal Length______________________\n")
print(df.sort_values("sepal_length"))
```

```python
print("__Average Sepal Length by Species__")
print(df.groupby("species")["sepal_length"].mean())
```

```python
df.to_csv("iris_updated.csv", index=False)
```
